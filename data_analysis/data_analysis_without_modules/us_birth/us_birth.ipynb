```python
# Analyzing data without numpy and pandas
import csv

f = open('US_births_1994-2003_CDC_NCHS.csv', 'r')
data = list(csv.reader(f, delimiter= ','))
data[:5]
```

```python
# header and data

data_header = data[0]
print(data_header)
data = data[1:]
data[:5]
```

```python
# making a dictionary for all the birth each month 
def month_birth(list_data):
    births_per_month = {}
    for value in list_data:
        if value[1] in births_per_month:
            births_per_month[value[1]] += int(value[4])
        else:
            births_per_month[value[1]] = int(value[4])
            
            
    return births_per_month
cdc_month_births = month_birth(data)
cdc_month_births
```

```python
# macking a dict for all the birth each day of week 

def dow_births(list_data):
    birth_per_week = {}
    for i in list_data:
        if i[3] in birth_per_week:
            birth_per_week[i[3]] += int(i[4])
        else:
            birth_per_week[i[3]] = int(i[4])
    return birth_per_week

cdc_day_birth = dow_births(data)
cdc_day_birth
```

```python
# building a func that responds for all the elelemts

def calc_counts(list_data, index):
    result = {}
    for i in list_data:
        if i[index] in result:
            result[i[index]] += int(i[4])
        else:
            result[i[index]] = int(i[4])
    return result

cdc_year_birth = calc_counts(data, 0)
print(cdc_year_birth)

cdc_dom_birth = calc_counts(data, 2)
cdc_dom_birth
```
