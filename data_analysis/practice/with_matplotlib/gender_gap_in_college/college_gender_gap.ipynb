```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

# Comparing across all degree categories

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics',
            'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism',
                'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education',
             'Agriculture', 'Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16, 20))
x_value = women_degrees['Year']

## Generate first column of line charts. STEM degrees.
for sp in range(0, 16, 3):
    if sp == 0:
        stem_index = sp
    else:
        stem_index = sp // 3
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(x_value, women_degrees[stem_cats[stem_index]], color=cb_dark_blue, label='Women', linewidth= 3)
    ax.plot(x_value, 100-women_degrees[stem_cats[stem_index]], color=cb_orange, label='Men', linewidth= 3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelbottom= False)
    
    ax.set_ylim(0,100)
    ax.set_yticks([0, 100])
    ax.set_xlim(1968, 2011)
    ax.set_title(stem_cats[stem_index])
    
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp ==0:
        ax.text(2005, 70, 'Women')
        ax.text(2007, 25, 'Men')
    if sp == 15:
        ax.text(2007, 77, 'Men')
        ax.text(2005, 22, 'Women')
        ax.tick_params(labelbottom= True)

#Generate second column of line charts. lib_art degrees
for sp in range(1, 14, 3):
    if sp == 1:
        lib_index = sp
    else:
        lib_index = sp // 3
    ax = fig.add_subplot(6,3, sp+1)
    
    ax.plot(x_value, women_degrees[lib_arts_cats[lib_index]], label='Women', color=cb_dark_blue, linewidth=3)
    ax.plot(x_value, 100 - women_degrees[lib_arts_cats[lib_index]], label= 'Men', color=cb_orange, linewidth=3)
    
    ax.set_title(lib_arts_cats[lib_index])
    
    ax.set_ylim(0,100)
    ax.set_yticks([1,100])
    ax.set_xlim(1968, 2011)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelbottom=False)
    
    if sp == 1:
        ax.text(2007, 40, 'Men')
        ax.text(2005, 60, 'Women')
    elif sp == 13:
        ax.text(2007, 56, 'Men')
        ax.text(2005, 40, 'Women')
        ax.tick_params(labelbottom=True)

# Generate third column of line charts. other degrees
for sp in range(2,18, 3):
    if sp == 2:
        other_index = sp
    else:
        other_index = sp // 3
    
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(x_value, women_degrees[other_cats[other_index]], label='Women', color=cb_dark_blue, linewidth=3)
    ax.plot(x_value, 100 - women_degrees[other_cats[other_index]], label='Men', color=cb_orange, linewidth=3)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelbottom= False)
    ax.set_title(other_cats[other_index])
    
    if sp == 2:
        ax.text(2007, 26, 'Men')
        ax.text(2005, 70, 'Women')
    elif sp == 17:
        ax.text(2007, 63, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom= True)

plt.savefig("gender_degrees.png")
    
    
```
