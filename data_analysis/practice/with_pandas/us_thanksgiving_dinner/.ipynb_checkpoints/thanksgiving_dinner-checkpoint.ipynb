```python
import pandas as pd

data = pd.read_csv('thanksgiving-data.csv', encoding='Latin-1')
data.head()
```

```python
data.columns
```

```python
# only getting the data from the people who celebrate Thanksgiving

data = data[data['Do you celebrate Thanksgiving?'] == 'Yes']
```

```python
# Visualize what people eat in Thanksgiving

what_is_eatten = data['What is typically the main dish at your Thanksgiving dinner?']
print(what_is_eatten.value_counts())
```

```python
eat_tofurkey = data[what_is_eatten == 'Tofurkey']
eat_tofurkey['Do you typically have gravy?']
```

```python
# Visualize all the people who eat apple, pumpkin, or pecan as dessert
apple_null = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'].isnull()
pumpkin_null = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin']
pecan_null = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan']

no_pies = data[apple_null & pumpkin_null & pecan_null]
print(no_pies)

#128 people do not eat any pies for Thanksgiving
```

```python
# More information about the age
import re
def change_age_to_int(single_val):
    p = re.compile('[0-9]+')
    m = p.findall(str(single_val))
    if not m:
        result = None
    else:   
        result = int(m[0])
    return result

data['int_age'] = data['Age'].apply(change_age_to_int)
        
print(data['int_age'].describe())

```

# Have to wirte more about the foundings 

```python
# More information about the income

def change_income_to_int(input_str):
    p = re.compile('[0-9,]+')
    m = p.findall(str(input_str))
    if not m:
        result = None
    else:
        if ',' in m[0]:
            m[0] = m[0].replace(',','')
        result = int(m[0])
    return result

data['int_income'] = data['How much total combined money did all members of your HOUSEHOLD earn last year?']\
.apply(change_income_to_int)

data['int_income'].describe()
```

# More info

```python
# Distance of travling and income

less_than_150000 = data[data['int_income'] <= 150000]
print(less_than_150000['How far will you travel for Thanksgiving?'].value_counts())

more_than_150000 = data[data['int_income'] > 150000]
print(more_than_150000['How far will you travel for Thanksgiving?'].value_counts())
```
