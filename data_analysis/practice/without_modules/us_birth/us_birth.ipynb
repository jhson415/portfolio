# Reading two datasets to list

```python
import csv

f = open('US_births_1994-2003_CDC_NCHS.csv', 'r')
data = list(csv.reader(f, delimiter= ','))
print(data[:5])

f = open('US_births_2000-2014_SSA.csv','r')
new_data = list(csv.reader(f, delimiter= ','))
new_data = new_data[1:]
```

# Combine two datasets together

```python
new_other_data = []
for v in new_data:
    if int(v[0]) > 2003:
        print(int(v[0]))
        new_other_data.append(v)
        
data = data + new_other_data
data[::200]
```

# Separating the header with the data

```python
header = data[0]
data = data[1:]
print(header)
data[:5]
```

# Preparing useful functions for analyzing data

## Returns the total birth for the needed area

```python
# Takes a list and the index number of the needed column as the argument
# Returns a dict type with the value of total birth for the needed date

def calc_counts(list_data, index):
    result = {}
    for i in list_data:
        if i[index] in result:
            result[i[index]] += int(i[4])
        else:
            result[i[index]] = int(i[4])
    return result
```

## Returns the day, week, month, year with the most and lest birth

```python
# Takes a dict that came out the calc_counts function as the argument
# Returns the max and min value in type str

def max_min_element(input_dict):
    max_value = None
    max_key = None
    
    for key, value in input_dict.items():
        if max_value == None or value > max_value:
            max_value = value
            max_key = key
    
    min_value = None
    min_key = None

    for key, value in input_dict.items():
        if min_value == None or min_value > value:
            min_value = value
            min_key = key
    
    result = "The max value is {0} with total of {1} birth, and the min value is {2} with total of {3} birth".format(max_key, max_value, min_key, min_value) 
    return(result)
```
