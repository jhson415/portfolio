# Drop out the rows the contain missing values 

```python
raw_data_count = len(recent_grads)
recent_grads.dropna(inplace=True)
cleanded_data_count = len(recent_grads)
missing_value = raw_data_count - cleanded_data_count

print('Raw data count: {0} \nCleaned data count: {1} \nRaw dataset have contained {2} missing value'\
      .format(raw_data_count, cleanded_data_count, missing_value))
```

# Using scatter plot to see the conection between the columns 

```python
# Rescaling the data
rescaled_recent_grads = recent_grads[['Total', 'Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']]

for column in rescaled_recent_grads.columns:
    recent_column = rescaled_recent_grads[column]
    rescaled_recent_grads[column] = (recent_column - recent_column.min()) / (recent_column.max() - recent_column.min())
```

```python
# See the relationship between the sample_size and median values of the earning 
rescaled_recent_grads.plot(x= 'Sample_size', y='Median', kind='scatter')
```

Most of the columns had low sample size, therefore it was hard to
line the conection for sample size, and Median value of the incomes.
Even with this condition, it is seen that the two columns do not have any conection.

```python
# See the relation between Sample size and unemployment_rate
rescaled_recent_grads.plot(x='Sample_size', y= 'Unemployment_rate', kind= 'scatter')
```

Same with the sample size, median scatter graph, it is hard to draw the conection since too many values of sample size stayes close to 0. The the unemplyment rate and the sample size do not seem to have any conection.

```python
# relation between full time jobs and median value for the income
rescaled_recent_grads.plot(x= 'Full_time', y='Median', kind='scatter')
```

It shows same result with the graphs above, no close conection 

```python
# Relationship between Share woman and unemployment rate
rescaled_recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

Unlike the other columns both columns unemployment rate, and share women seems to have a fairly spread values, with normal variation. When it comes to the relationship between the two series, they do not have conection 

```python
# Relationship between number of men working and the median value of the income
rescaled_recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
# Relationship between number of women working and the median value of the income
rescaled_recent_grads.plot(x='Women', y='Median', kind='scatter')
```

Both men and women numbers in the major did not show and kind of conection with the median income, but these two graph slightly shows that majors with more men have higher income.

```python
# Conection between median income and share women
rescaled_recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
```

This graph proves the idea that came out from the previous graph. When the women share gets higher it more likely to have a lower median income for the workers.

```python
# Relationthip between total people in the major and the median income
rescaled_recent_grads.plot(x='Total', y='Median', kind='scatter')
```

A interesting point came out from this graph, when the total people in the major was low, it have a great SD. The rescent grads got either very high or low income. When the total number of the students grew, the median income of the students where not that high or not that low either.
