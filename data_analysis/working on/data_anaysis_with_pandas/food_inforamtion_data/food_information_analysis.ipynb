## Normalize all the column data to mg

```python
import re
# Making a list of the column labels 
column_names = data.columns
# Creating a new data frame
normalized_data = pd.DataFrame([])

# Inputing the normalized data to the new data frame
for column in column_names:
    p = re.compile('[a-zA-Z_]*')
    new_label = p.findall(column)
    if data[column].dtypes == 'float64' or column == 'Energ_Kcal':
        normalized_data[new_label[0]] = (data[column] - data[column].min()) / (data[column].max() - data[column].min())
    else:
        normalized_data[new_label[0]] = data[column]
        
normalized_data
```

# Things that can be analyzed through out this dataset

### Return the top 5 foods of each columns 

```python
for column in column_names[2:]:
    top_five_df = pd.DataFrame()
    top_five_df['name'] = data['Shrt_Desc']
    top_five_df[column] = data[column]
    top_five_df.sort_values([column], inplace=True, ascending=False)
    print(top_five_df[:5])
```
